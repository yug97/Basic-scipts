### Importing Packages and Loading Data

```python
%matplotlib inline
import pandas as pd
import statsmodels.api as sm
import warnings
import itertools
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import pandas_profiling as pp
```

```python
df = pd.DataFrame.from_records(sm.datasets.co2.load().data)
df['date'] = df.date.apply(lambda x : x.decode('utf-8'))
df['date'] = pd.to_datetime(df.date,format='%Y%m%d')
df.set_index('date',inplace = True)
df.head(5)
```

```python
y = df['co2'].resample('MS').mean()
y.head(5)
```

```python
y.isnull().sum()
```

```python
y = y.fillna(y.bfill())
y.isnull().sum()
```

```python
y.plot(figsize=(12,9))
```

### Parameter Selection for the ARIMA Time Series Model

```python
p = d = q = range(0,2) # can take values 0, 1
pdq = list(itertools.product(p,d,q))# All combination of triplets 
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in pdq] #All combinations of seasonal p,q,d trpletsb
```

```python
warnings.filterwarnings('ignore')
Aic = []
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(y,order=param,
                                       seasonal_order=param_seasonal,
                                       enforce_invertibility=False,
                                       enforce_stationarity=False)
            result = mod.fit()
        
            #print('SARIMAX {} x {} - AIC {}'.format(param,param_seasonal,result.aic))
            Aic.append((param,param_seasonal,result.aic))
        except:
            continue
```

```python
import operator
best_param = sorted(Aic, key= operator.itemgetter(2))
best_param[0]
```

### Fitting an ARIMA Time Series Model

```python
model = sm.tsa.SARIMAX(endog = y,
                      order = (1,1,1),
                      seasonal_order= (1,1,1,12),
                      enforce_invertibility = False,
                      enforce_stationarity = False)
res = model.fit()

print(res.summary())
```

```python
res.plot_diagnostics(figsize=(15,12))
```

### Validating Forecasts (dynamic = False)

```python
pred = res.get_prediction(start = pd.to_datetime('1998-01-01'),dynamic = False)
pred_ci = pred.conf_int()
pred_ci.head(5)
```

```python
ax = y['1990':].plot(label = 'observed',figsize=(15,12))
pred.predicted_mean.plot(ax=ax, label = 'One-step ahead Forecast', alpha=.7)
ax.fill_between(pred_ci.index,
               pred_ci.iloc[:,0],
               pred_ci.iloc[:,1], color = 'k', alpha = 0.3)

ax.set_xlabel('Date')
ax.set_ylabel('CO2 Levels')
plt.legend()
```

```python
### MSE
y_forecasted = pred.predicted_mean
y_true = y['1998-01-01':]

mse = ((y_forecasted - y_true)**2).mean()
print('MSE : {}'.format(round(mse,2)))
```

### Validating Forecasts (dynamic = False)

```python
pred_dyna = res.get_prediction(start = pd.to_datetime('1998-01-01'), dynamic = True)
pred_dyna_ci = pred_dyna.conf_int()
ax = y['1990':].plot(label = 'observed',figsize=(15,12))
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7)

ax.fill_between(pred_dyna_ci.index,
               pred_dyna_ci.iloc[:,0],
               pred_dyna_ci.iloc[:,1], color = 'k' , alpha = 0.2)
ax.set_xlabel('Date')
ax.set_ylabel('CO2 Levels')
plt.legend()
```

```python
#MSE

y_forecasted_dyna = pred_dyna.predicted_mean
y_true_dyna = y['1998-01-01':]

mse_dy = ((y_forecasted_dyna - y_true_dyna)**2).mean()
print('MSE (dynamic): {}'.format(mse_dy))
```

### Producing and Visualizing Forecasts

```python
pred_uc = res.get_forecast(steps = 500)
pred_uc_ci = pred_uc.conf_int()

ax = y.plot(label='observed',figsize=(25,20))
pred_uc.predicted_mean.plot(ax = ax,  label = 'Forecast')
ax.fill_between(pred_uc_ci.index,
               pred_uc_ci.iloc[:,0],
               pred_uc_ci.iloc[:,1], color = 'k', alpha = 0.2)
ax.set_xlabel('Date')
ax.set_ylabel('CO2 Levels')

plt.legend()
```
